## Data Cleaning

```python
import pandas as pd

file1 = "../data/results_1_73bit.tsv"
file2 = "../data/results_2_22bit.tsv"
file3 = "../data/results_2_51bit.tsv"

df1 = pd.read_csv(file1, sep="\t")
df2 = pd.read_csv(file2, sep="\t")
df3 = pd.read_csv(file3, sep="\t")

df1["Quantization Level (Bits)"] = 1.73
df2["Quantization Level (Bits)"] = 2.22
df3["Quantization Level (Bits)"] = 2.51

merged_df = pd.concat([df1, df2, df3], ignore_index=True)

rename_map = {
    'sample_index': 'Data Sample Index',
    'question': 'Question',
    'reference_answer': 'Reference Answer',
    'generated_answer': 'Generated Answer',
    'rougeL': 'ROUGE-L',
    'bertscore_f1': 'BERTScore-F1',
    'recall@3': 'Recall@3',
    'ndcg@3': 'NDCG@3',
    'recall@5': 'Recall@5',
    'ndcg@5': 'NDCG@5',
    'peak_cpu': 'Peak CPU (%)',
    'peak_mem': 'Peak RAM (GiB)',
    'peak_vram': 'Peak VRAM (GiB)',
    'load_time': 'Load Time (min)',
    'prompt_eval_time': 'Prompt Eval Time (min)',
    'eval_time': 'Eval Time (min)',
    'total_time': 'Total Time (min)',
    'sampling_time': 'Sampling Time (min)'
}

merged_df.rename(columns=rename_map, inplace=True)

# Numerical cols to float
non_numerical_cols = ['Question', 'Reference Answer', 'Generated Answer']
for col in [c for c in merged_df.columns if c not in non_numerical_cols]:
    merged_df[col] = pd.to_numeric(merged_df[col], errors='coerce')

# MiB to GiB
merged_df['Peak RAM (GiB)'] = merged_df['Peak RAM (GiB)'] / 1024
merged_df['Peak VRAM (GiB)'] = merged_df['Peak VRAM (GiB)'] / 1024

# Time to Mins
time_columns = [
    'Load Time (min)', 'Prompt Eval Time (min)', 'Eval Time (min)',
    'Total Time (min)', 'Sampling Time (min)'
]

for col in time_columns:
    merged_df[col] = merged_df[col].apply(lambda x: x / 60000 if x > 600 else x)

# These cols don't depend on LLM output and therefore don't exhibit variability between quant levels
cols_to_drop = ['Recall@3', 'NDCG@3', 'Recall@5', 'NDCG@5']
merged_df = merged_df.drop(columns=cols_to_drop)

merged_df.to_csv('results_cleaned.csv', index=False)
```

## Plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Group by quantization level and compute means
grouped = merged_df.groupby("Quantization Level (Bits)").mean(numeric_only=True)

# Plot: Quantization vs Generation Quality (Zoomed)
plt.figure(figsize=(8, 6))
plt.plot(grouped.index, grouped["ROUGE-L"], marker="o", label="ROUGE-L")
plt.plot(grouped.index, grouped["BERTScore-F1"], marker="s", label="BERTScore-F1")
plt.title("Quantization vs Generation Quality")
plt.xlabel("Quantization Level (Bits)")
plt.ylabel("Score")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("../plots/quant_vs_rouge_bert.png", dpi=300, bbox_inches='tight')
```

```python
# Plot: Quantization vs Inference Time (Total Time)
offloaded_layers = {
    1.73: 4,
    2.22: 3,
    2.51: 2
}

quant_levels = grouped.index
total_time = grouped["Total Time (min)"]
prompt_eval_time = grouped["Prompt Eval Time (min)"]
eval_time = grouped["Eval Time (min)"]
offloaded = [offloaded_layers[q] for q in quant_levels]

# Plot the time-based metrics
plt.figure(figsize=(8, 6))
plt.plot(quant_levels, total_time, marker='o', label="Total Time")
plt.plot(quant_levels, prompt_eval_time, marker='s', label="Prompt Eval Time")
plt.plot(quant_levels, eval_time, marker='^', label="Eval Time")

# Annotate with number of offloaded layers
for x, y, layers in zip(quant_levels, total_time, offloaded):
    plt.text(x, y + 0.1, f"{layers} layers", ha='center', fontsize=9, color='red')

plt.title("Quantization Level vs Inference Time Components\n(with Number of GPU-Offloaded Layers)")
plt.xlabel("Quantization Level (Bits)")
plt.ylabel("Time (minutes)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("../plots/quant_vs_time.png", dpi=300, bbox_inches='tight')
```

```python
grouped_resources = merged_df.groupby("Quantization Level (Bits)").mean(numeric_only=True)

# Plot
plt.figure(figsize=(8, 6))
plt.plot(grouped_resources.index, grouped_resources["Peak VRAM (GiB)"], marker='o', label="Peak VRAM (GiB)")
plt.plot(grouped_resources.index, grouped_resources["Peak RAM (GiB)"], marker='s', label="Peak RAM (GiB)")
plt.plot(grouped_resources.index, grouped_resources["Peak CPU (%)"], marker='^', label="Peak CPU (%)")

plt.title("Quantization Level vs System Resource Usage")
plt.xlabel("Quantization Level (Bits)")
plt.ylabel("Usage")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("../plots/quant_vs_resources.png", dpi=300, bbox_inches='tight')
```

```python
# Plot with annotations: Total Time vs. ROUGE-L
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(
    data=merged_df,
    x="Total Time (min)",
    y="ROUGE-L",
    hue="Quantization Level (Bits)",
    palette="viridis",
    alpha=0.6
)

# Add mean point for each quantization level
grouped = merged_df.groupby("Quantization Level (Bits)").agg({
    "Total Time (min)": "mean",
    "ROUGE-L": "mean"
}).reset_index()

for _, row in grouped.iterrows():
    plt.scatter(row["Total Time (min)"], row["ROUGE-L"], color="black", marker="X", s=100, label="Mean per Quantization Level")
    plt.text(row["Total Time (min)"], row["ROUGE-L"] + 0.001,
             f'{row["Quantization Level (Bits)"]}b', ha='center', fontsize=9)

# Make legend
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys())

plt.title("Efficiency vs Accuracy (ROUGE-L)")
plt.xlabel("Total Time (min)")
plt.ylabel("ROUGE-L")
plt.grid(True)
plt.tight_layout()
plt.savefig("../plots/efficiency_vs_rouge.png", dpi=300, bbox_inches='tight')
```

```python
# Set plot style
sns.set(style="whitegrid")

# Scatter plot: Total Time vs. BERTScore-F1
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(
    data=merged_df,
    x="Total Time (min)",
    y="BERTScore-F1",
    hue="Quantization Level (Bits)",
    palette="viridis",
    alpha=0.7,
    edgecolor="k"
)

# Add average markers per quantization level
avg_points = merged_df.groupby("Quantization Level (Bits)").agg({
    "Total Time (min)": "mean",
    "BERTScore-F1": "mean"
}).reset_index()

# Plot and annotate mean points
for _, row in avg_points.iterrows():
    plt.scatter(
        row["Total Time (min)"],
        row["BERTScore-F1"],
        color="black",
        marker="X",
        s=100,
        label="Mean per Quantization Level"
    )
    plt.text(
        row["Total Time (min)"],
        row["BERTScore-F1"] + 0.0005,
        f'{row["Quantization Level (Bits)"]}b',
        ha='center',
        fontsize=9,
        color="black"
    )

# Deduplicate legend
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys(), title="Quantization Level (Bits)")

plt.title("Efficiency vs. Generation Quality (BERTScore-F1)")
plt.xlabel("Total Inference Time (min)")
plt.ylabel("BERTScore-F1")
plt.tight_layout()
plt.savefig("../plots/efficiency_vs_bert.png", dpi=300, bbox_inches='tight')
```

```python
df = merged_df

# Identify numeric columns and convert
non_numeric = ['Data Sample Index', 'Question', 'Reference Answer', 'Generated Answer', 'Quantization Level (Bits)']
numeric_cols = df.columns.drop(non_numeric, errors='ignore')
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')

# Compute summary statistics
summary_mean = df.groupby("Quantization Level (Bits)")[numeric_cols].mean()
summary_std = df.groupby("Quantization Level (Bits)")[numeric_cols].std()

# Highlight function for means
def highlight_performance(row):
    return ['background-color: lightgreen' if (
        ('ROUGE' in col or 'BERT' in col or 'Recall' in col or 'NDCG' in col) and row[col] == summary_mean[col].max()
        or ('Time' in col or 'CPU' in col or 'RAM' in col or 'VRAM' in col) and row[col] == summary_mean[col].min()
    ) else '' for col in row.index]

# Apply styling
styled_mean = summary_mean.style.apply(highlight_performance, axis=1)
styled_std = summary_std.style.background_gradient(cmap='Purples')
```
